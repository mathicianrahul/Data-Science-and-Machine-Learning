```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("housing.csv")
df.head()
```

```python
import numpy as np
df["income_cat"]=pd.cut(df["median_income"],bins=[0,1.5,3.0,4.5,6,np.inf],labels=["1","2","3","4","5"])
df.head()
```

```python
import matplotlib.pyplot as plt
df['income_cat'].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Category Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number Of Instance")
plt.show()
```

```python
#Now Let's saperate the equal Category data into the both training and testing dataset
from sklearn.model_selection import StratifiedShuffleSplit

split=StratifiedShuffleSplit(n_splits=1 , test_size=0.2 , random_state=42)
for train_index ,test_index in split.split(df,df["income_cat"]):
    strat_train_set=df.loc[train_index]
    strat_test_set=df.loc[test_index]
```

```python
#Now let's Check the train, test graph (IS IT BOTH SAME) IF YES THEN THE EVERY CATEGORY DATA ARE PRESENT IN BOTH DATASET
import matplotlib.pyplot as plt
strat_train_set['income_cat'].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Category Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number Of Instance")
plt.show()
```

```python
strat_test_set["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Category Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number Of Instance")
plt.show()
```
